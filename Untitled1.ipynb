```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df=pd.DataFrame(pd.read_excel(r"/content/Minor Project Data set (Stock Price Prediction).xlsx"))

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set the date as the index
df = df.set_index('Date')

# Calculate the daily return
df['Return'] = df['Adj Close'].pct_change()

# Calculate the cumulative return
df['Cumulative Return'] = (1 + df['Return']).cumprod()

# Calculate the average daily return
df['Average Daily Return'] = df['Return'].mean()

# Calculate the standard deviation of daily returns
df['Standard Deviation'] = df['Return'].std()

# Calculate the Sharpe Ratio
df['Sharpe Ratio'] = df['Average Daily Return'] / df['Standard Deviation']

# Plot the cumulative return
plt.figure(figsize=(10, 5))
plt.plot(df['Cumulative Return'], label='Cumulative Return')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.title('Cumulative Return')
plt.legend()
plt.show()

# Plot the daily returns
plt.figure(figsize=(10, 5))
plt.hist(df['Return'], bins=50)
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.title('Histogram of Daily Returns')
plt.show()

# Plot the correlation matrix
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
